Data Cleaning

```python
# 1. Clean column names
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
      .str.replace('-', '_')
)
df.columns
```

```python
# Replace 'N/A' with actual NaN and convert to numeric
df['regents_exam'] = df['regents_exam'].replace(' ', np.nan)
df['number_scoring_cr'] = df['number_scoring_cr'].replace('s', np.nan)
df['number_scoring_cr'] = pd.to_numeric(df['number_scoring_cr'], errors='coerce')
df['percent_scoring_cr'] = df['percent_scoring_cr'].replace('s', np.nan)
df['percent_scoring_cr'] = pd.to_numeric(df['percent_scoring_cr'], errors='coerce')
df['total_tested'] = pd.to_numeric(df['total_tested'], errors='coerce')
df['mean_score'] = pd.to_numeric(df['mean_score'], errors='coerce')
df['number_scoring_below_65'] = pd.to_numeric(df['number_scoring_below_65'], errors='coerce')
df['percent_scoring_below_65'] = pd.to_numeric(df['percent_scoring_below_65'], errors='coerce')
df['number_scoring_65_or_above'] = pd.to_numeric(df['number_scoring_65_or_above'], errors='coerce')
df['percent_scoring_65_or_above'] = pd.to_numeric(df['percent_scoring_65_or_above'], errors='coerce')
df['number_scoring_80_or_above'] = pd.to_numeric(df['number_scoring_80_or_above'], errors='coerce')
df['percent_scoring_80_or_above'] = pd.to_numeric(df['percent_scoring_80_or_above'], errors='coerce')
df[['regents_exam', 'number_scoring_cr', 'percent_scoring_cr', 'total_tested', 'mean_score', 'number_scoring_below_65', 'percent_scoring_below_65','number_scoring_65_or_above','percent_scoring_65_or_above','number_scoring_80_or_above','percent_scoring_80_or_above']].info()
```

Converting Columns to Categorical

```python
# Re-apply column cleaning to ensure consistency
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
      .str.replace('-', '_')
)

categorical_cols = [
    'school_dbn', 'school_name', 'school_level', 'regents_exam', 'year']

for col in categorical_cols:
    df[col] = df[col].astype('category')

df[categorical_cols].dtypes
```

Descriptive Statistics - Numeric Column

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
numeric_df.describe()
```

```python
# Filter Stuyvesant HS
stuy = df[df["school_name"].str.contains("Stuyvesant", case=False, na=False)]

# Filter Common Core Algebra exam score
stuy_alg = stuy[stuy["regents_exam"].str.contains("Common Core Algebra", case=False, na=False)]

# Stuyvesant mean score by year
stuy_means = stuy_alg.groupby("year")["mean_score"].mean()
print(stuy_means)
```

```python
# Filter Queens high schools
df["Borough"] = df["school_dbn"].str[2]  # 3rd character

queens_hs = df[
    (df["Borough"] == "Q") &
    (df["school_level"].str.contains("High", case=False, na=False))
]

# Filter Common Core Algebra exam score
queens_alg = queens_hs[queens_hs["regents_exam"].str.contains("Common Core Algebra", case=False, na=False)]

# Queens average mean score by year
queens_means = queens_alg.groupby("year")["mean_score"].mean()
print(queens_means)
```

```python
import matplotlib.pyplot as plt

years = sorted(set(stuy_means.index).intersection(queens_means.index))

plt.figure(figsize=(8,5))

plt.plot(years, [queens_means[y] for y in years], marker='o', label="Queens High Schools Avg")
plt.plot(years, [stuy_means[y] for y in years], marker='o', linestyle="--", label="Stuyvesant High School")

plt.title("Common Core Algebra  Mean Score Comparison\nStuyvesant HS vs Queens High Schools")
plt.xlabel("Year")
plt.ylabel("Mean Score")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

df=pd.read_csv("https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv")

# Re-apply column cleaning
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
      .str.replace('-', '_')
)

# Ensure 'percent_scoring_65_or_above' is numeric
df['percent_scoring_65_or_above'] = df['percent_scoring_65_or_above'].replace('s', np.nan)
df['percent_scoring_65_or_above'] = pd.to_numeric(df['percent_scoring_65_or_above'], errors='coerce')

# Filter Stuyvesant HS
stuy = df[df["school_name"].str.contains("Stuyvesant", case=False, na=False)]

# Focus on Common Core Algebra exam
exam = "Common Core Algebra"
stuy_pass = stuy[stuy["regents_exam"].str.contains(exam, case=False, na=False)]

# Pass Rates Group by Year
stuy_rates = stuy_pass.groupby("year")["percent_scoring_65_or_above"].mean()
print(stuy_rates)
```

```python
# Filter Queens high schools
df["Borough"] = df["school_dbn"].str[2]  # 3rd character

queens_hs = df[
    (df["Borough"] == "Q") &
    (df["school_level"].str.contains("High", case=False, na=False))
]

# Focus on Common Core Algebra Exam
queens_pass = queens_hs[queens_hs["regents_exam"].str.contains("Common Core Algebra", case=False, na=False)]

# Group by Year
queens_rates = queens_pass.groupby("year")["percent_scoring_65_or_above"].mean()
print(queens_rates)
```

```python
# Create bar chart
years = sorted(set(stuy_rates.index).intersection(queens_rates.index))
x = range(len(years))
width = 0.35

plt.figure(figsize=(8,5))

plt.bar([i - width/2 for i in x], [stuy_rates[y] for y in years], width, label="Stuyvesant HS")
plt.bar([i + width/2 for i in x], [queens_rates[y] for y in years], width, label="Queens HS Avg")

plt.xticks(x, years)
plt.ylabel("Percent Scoring 65 or Above")
plt.title("Pass Rates: Stuyvesant HS vs Queens High Schools\n(Common Core Algebra)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
Stuyvesant High School consistently outperforms the average Queens High Schools on the Common Core Algebra Regents exam.
However, Stuyvesant experienced a noticeable decline in 2017, nearly matching the borough-wide average.
Overall, it maintains a strong record, but the 2017 decline requires closer examination.
```
